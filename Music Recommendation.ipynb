```python
song_df_1 = pd.read_csv('archive/User Listening History.csv')
print (song_df_1.shape)
print (song_df_1.head())
```

```python
song_df_2 = pd.read_csv('archive/Music Info.csv')
print (song_df_2.shape)
display  (song_df_2.head())
```

```python
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['track_id']), on='track_id', how='left')
display (song_df.shape)
display (song_df.head())
```

```python
print(len(song_df_1), len(song_df_2))
```

```python
len(song_df)
```

```python
song_df = song_df.head(50000)
song_df.shape
```

```python
song_df['song'] = song_df['name']+' - '+song_df['artist']
song_df.head()
```

```python
user=song_df["user_id"].drop_duplicates()
user.index = range(1, 1 + len(user))
user.iloc[0]
user

```

```python
song_grouped = song_df.groupby(['song']).agg({'playcount':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['playcount'].sum()
grouped_sum
```

```python
song_grouped['percentage'] = (song_grouped['playcount'] / grouped_sum ) * 100
song_grouped.sort_values(['playcount', 'song'], ascending=[0,1])
```

```python
pr = Recommenders.popularity_recommender_py()
pr.create(song_df, 'user_id', 'song')
rec=pr.recommend(song_df['user_id'])
```

```python
rec
```

```python
for i in rec['song']:
    song_details = song_df[song_df['song'] == i] 
    song_name = song_details.iloc[0]['name']
    artist_name = song_details.iloc[0]['artist']
    link = song_details.iloc[0]['spotify_preview_url']
    print(song_name,artist_name,link)
```

```python
df=rec
df=df.drop(columns=['user_id'])
df.reset_index(drop=True)
```
